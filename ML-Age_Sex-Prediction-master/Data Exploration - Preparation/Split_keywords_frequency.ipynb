```python
#fonction pour separer les mots clés dans un dataset afin de pouvoir les manipuler 
def splitwords_to_df(linekeyword):
    words = []
    frequencies = []
    newlist = []
    listkeywords = linekeyword.split(";")
    for n in listkeywords:
        newlist = newlist + n.split(':')
    for n in newlist:
        if newlist.index(n)%2 == 0:
            words.append(n)
        else:
            # on force le passage en int sinon les frequences sont en string 
            frequencies.append(int(n))
    finaldf = pd.DataFrame({"keywords": words, "frequency": frequencies}) 
    # finaltuple = tuple(map(lambda x, y: (x,y) , words, frequencies)) test avec un tupple en output
    return finaldf
```

# Final function to split the keywords and the frequency by id

```python
def tidy_split(df, column, sep=';', keep=False):
    indexes = list()
    new_values = list()
    new_freq = list()
    df = df.dropna(subset=[column])
    for i, presplit in enumerate(df[column].astype(str)):
        values = presplit.split(sep)
        if keep and len(values) > 1:
            indexes.append(i)
            new_values.append(presplit)
        for value in values:
            indexes.append(i)
            new_values.append(value.split(":")[0])
            new_freq.append(value.split(":")[1]) if len(value.split(":")) > 1 else  new_freq.append(0)
    new_df = df.iloc[indexes, :].copy()
    new_df[column] = new_values
    new_df['frequency'] = new_freq
    return new_df
```
